# Visualizer

```python
# import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# helper function for visualization

def visualizer(x, y, plot_type, title, xlabel, ylabel, rotation=False, rotation_value=60, figsize=(13,7)):
    '''helper visualizer'''
    
    plt.figure(figsize=figsize)
    
    if plot_type == 'bar':  
        sns.barplot(x=x, y=y)
    elif plot_type == 'count':  
        sns.countplot(x)
    elif plot_type == 'regression':  
        sns.regplot(x=x,y=y)
    elif plot_type == 'scatter':
        plt.scatter(x,y)

    plt.title(title, fontsize=16)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    
    if rotation:
        plt.xticks(rotation=rotation_value)
    
    plt.show()
    
```

```python
# read csv
df = pd.read_csv('visualizer_app_store_1.csv')
```

### Visuals

```python
# top 10 apps by rating

rating_df = df.sort_values('rating_count_tot', ascending=False)
rating_df[:10]
```

```python
visualizer(rating_df[:10].track_name, rating_df[:10].rating_count_tot, 'bar', 'TOP 10 APPS BY RATING',
          'APP NAME', 'RATING COUNT (TOTAL)', True, -60)
```

```python
# top 10 apps by app size

app_size_df = df.sort_values('size_bytes', ascending=False)
app_size_df.size_bytes /= 1024*1024*1024 #Conversion from Bytes to GigaBytes
app_size_df[:10]
```

```python
visualizer(app_size_df[:10].track_name, app_size_df[:10].size_bytes, 'bar', 
           'TOP 10 APPS BY SIZE', 'APPLICATION', 'SIZE (GBs)', True, -60)
```

```python
# top 10 apps by app prize

app_price_df = df.sort_values('price', ascending=False)
app_price_df[:10]
```

```python
visualizer(app_price_df[:10].track_name, app_price_df[:10].price, 'bar', 
           'TOP 10 APPS BY PRICE', 'APPLICATION', 'PRICE $', True)
```

```python
correlations = df.corr()
correlations
```

```python
# Generate a mask for the upper triangle
mask = np.zeros_like(correlations, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True
```

```python
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(17, 13))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(correlations, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
