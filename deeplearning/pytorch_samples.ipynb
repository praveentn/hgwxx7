# PyTorch

```python
import torch
```

```python
x = torch.ones(2, 2, requires_grad=True)
print(x)
```

```python
y = x + 2
print(y)
```

```python
print(y.grad_fn)
```

```python
z = y * y * 3
out = z.mean()

print(z, out)
```

```python
a = torch.randn(2, 2)
a = ((a * 3) / (a - 1))
print(a.requires_grad)
```

```python
a.requires_grad_(True)
print(a.requires_grad)
```

```python
b = (a * a).sum()
print(b.grad_fn)
```

```python
x = torch.ones(2, 2, requires_grad=False)
print(x)
```

```python
y = x + 2
print(y)
```

```python
x = torch.randn(3, requires_grad=True)

y = x * 2
while y.data.norm() < 1000:
    y = y * 2

print(y)
```

```python
v = torch.tensor([0.1, 1.0, 0.0001], dtype=torch.float)
y.backward(v)

print(x.grad)
```
