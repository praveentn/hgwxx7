```python
n = 6
qc = QuantumCircuit(n)
qc.h(0)
for i in range(n-1):
    qc.cx(0, i+1)
qc.x(1)

sv = Statevector.from_instruction(qc)
plot_state_qsphere(sv, show_state_labels=False)
```
